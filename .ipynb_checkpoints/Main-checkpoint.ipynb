# Regions of the stellar initial mass function

# Data Analysis

```python
import numpy as np

import astropy
import astropy.units as u
from astropy.io import ascii

from matplotlib import pyplot as plt

import pandas as pd

import seaborn as sns

plt.rcParams['text.usetex'] = True
```

```python
data = ascii.read("table5.dat", readme="ReadMe.txt")
```

```python
df = data.to_pandas()
```

```python
df
```

```python
oRA = df["oRA"]*u.arcsec
oDE = df["oDE"]*u.arcsec
```

```python
fig, ax = plt.subplots()

ax.scatter(oRA,oDE,s=0.5,cmap='grey',c=df["M(RL)"])
```

## Present and Initial Mass

```python
filtered_initial_mass_RL_df = df[df["Mi(RL)"] != -999]
sns.kdeplot(filtered_initial_mass_RL_df["Mi(RL)"])

filtered_initial_mass_N_df = df[df["Mi(N)"] != -99]
sns.kdeplot(filtered_initial_mass_N_df["Mi(N)"])

plt.title("KDE for the Initial Mass")
plt.xlabel(r"M_{\odot}")
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)

filtered_initial_mass_RL_df = df[df["Mi(RL)"] != -999]
sns.kdeplot(filtered_initial_mass_RL_df["Mi(RL)"])
ax1.set_title("")

filtered_initial_mass_N_df = df[df["Mi(N)"] != -99]
sns.kdeplot(filtered_initial_mass_N_df["Mi(N)"])
```
