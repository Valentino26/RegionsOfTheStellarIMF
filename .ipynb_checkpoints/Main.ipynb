# Regions of the stellar initial mass function

### Introduction
The stellar **initial mass function** (IMF) is one of the cornerstone concepts in modern astrophysics. Defined as the distribution of stellar masses at the time of their formation, the IMF has profound implications for star formation, galaxy evolution, and the synthesis of heavy elements. The IMF determines how mass is partitioned among stars of different masses in a star-forming region and acts as a bridge connecting microphysical processes in star formation to macroscopic phenomena in galaxies and the universe.

Mathematically, the IMF is commonly expressed as:

$$
\xi(m) = \frac{dN}{dm} \quad \text{or} \quad \frac{dN}{d \log{m}} \propto \left(\frac{m}{M_{\odot}}\right) ^{\alpha + 1},
$$

where $\xi(m)$ represents the number of stars forming per unit mass interval $ dm $ at a mass $ m $. The IMF is often approximated by a power law or piecewise power-law function, such as the Salpeter (1955) IMF:

$$
\xi(m) \propto m^{-\alpha}, \quad \text{for} \quad m \geq 0.5 \mathrm{M_\odot},
$$

with $\alpha = 2.35 $ for higher-mass stars. Extensions to the Salpeter IMF, such as the Kroupa (2001) or Chabrier (2003) forms, include modifications to account for the flattening at lower stellar masses.

The importance of the IMF arises from its pervasive role in astrophysical calculations. For example:

1. **Star Formation Rates**: The IMF determines the relationship between the observed light from young stellar populations and the total mass of stars formed.
2. **Chemical Enrichment**: Massive stars produce most of the universe's heavy elements, and the IMF controls the fraction of stellar mass that ends its life in supernova explosions.
3. **Galactic Dynamics**: The IMF impacts the mass-to-light ratios used to estimate the dark matter content in galaxies.
4. **Exoplanet Studies**: The frequency of low-mass stars (M-dwarfs) in the IMF influences the probability of finding habitable planets.

## Working with the Arches Cluster Infrared Photometry, Extinction, and Masses

This daataset provides comprhensive photometric, extinction, and mass information (initial mass and present mass for two different extinciton laws (N) and (RL)) for stars in the **Arches Cluster**, a massive young starburst cluster located near the Galactic Center. The dataset is critical for studying the dynamical mass segregation, **stellar mass function**, and the influence of the extinction laws (Rieke and Lebofsky Extinction Law (RL-EL)) and (Nishyama et al. Extinction Law (N-EL)) on stellar properties in extreme environments.

The Dataset provides the following characteristics of stars:
1. **Coordinates**:
   - **Declination Offset [arcsec]**: Offset in declination from the reference star (J2000).
   - **Right Ascension Offset [arcsec]**: Offset in right ascension from the reference star (J2000).

2. **Extinction**:
   - **$ A_{Ks} $ (RL)**: $ Ks $-band extinction derived using the Rieke and Lebofsky (RL) extinction law.
   - **$ A_{Ks} $ (N)**: $ Ks $-band extinction derived using the Nishiyama (N) extinction law.

3. **Masses**:
   - **$ M $ (RL)**: Present-day mass derived using the RL extinction law.
   - **$ M $ (N)**: Present-day mass derived using the N extinction law.
   - **$ M_i $ (RL)**: Initial mass derived using the RL extinction law.
   - **$ M_i $ (N)**: Initial mass derived using the N extinction law.

4. **Photometry**:
   - **$ J $-band Brightness ($ J_{mag} $)**: Apparent brightness in the $ J $-band.
   - **$ H $-band Brightness ($ H_{mag} $)**: Apparent brightness in the $ H $-band.
   - **$ Ks $-band Brightness ($ Ks_{mag} $)**: Apparent brightness in the $ Ks $-band.

5. **Estimated Extinction**:
   - **$ A(Ks) $**: Estimated $ Ks $-band extinction value.

6. **Cluster Properties**:
   - **$ F $**: Field in which the source is located (1â€“5).
   - **$ Mm $**: Cluster membership indicator:
     - $ 1 $: Cluster member.
     - $ 0 $: Non-member or field star.

### Present and Initial Mass Distributions (KDE Plots of Mass Distributions (Comparing Extinction Laws))

The Dataset shows some physically impossible numbers such as $-999 M_\odot$, these numbers have to be excluded/filtered out as they are delibaretly set to these values, depending on whether there was data available for said star or not. Essentially we can characterize them as NaN.

To display the data we used a **(Gaussian) Kernel Density Estimate Plot**.

A kernel density plot in Seaborn is a way to estimate the PDF of a random variable based on a finite sample of data.
Seaborn does this by using a Gaussian Kernel $K(u) = \frac{1}{\sqrt{2\pi}}e^{-u^2/2}$ and applies it to the kernel density estimator

$$
\hat{f}(x) = \frac{1}{nh}\sum_{i=1}^n K\left( \frac{x - x_i}{h} \right)
$$

where $n$ is the number of data points in the sample, $h>0$ is the Bandwith(smoothing parameter) that controls the width of the kernel and K is the mentioned Kernel.

```python
# initialize  figure
fig, ax = plt.subplots()

# exclude all dataframe entries with Mi(RL) values of -999
filtered_initial_mass_RL_df = df[df["Mi(RL)"] != -999]
sns.kdeplot(filtered_initial_mass_RL_df["Mi(RL)"],ax=ax,label="PDF Mi(R-EL)")

# exclude dataframe entries with Mi(N) valeus of -99
filtered_initial_mass_N_df = df[df["Mi(N)"] != -99]
sns.kdeplot(filtered_initial_mass_N_df["Mi(N)"],color='orange',ax=ax,label="PDF Mi(N-EL)")

ax.set_title("KDE for the Initial Mass")
ax.set_xlabel(r"M_{\odot}")
ax.legend()
```

```python
stats_combined(["Mi(RL)","Mi(N)"],filtered_initial_mass_RL_df["Mi(RL)"],filtered_initial_mass_RL_df["Mi(N)"])
```

This figure shows the kernel density estimation (KDE) of the initial stellar masses derived using two different extinction laws: Rieke and Lebofsky (RL-EL) and Nishiyama (N-EL).

1. **Extinction Law Impact**:
   - The **blue curve (RL-EL)** and the **orange curve (N-EL)** demonstrate differences in the mass distribution, particularly at lower masses.
   - The **N-EL distribution (orange)** shows a sharper peak at low masses ($\sim 10 \, M_{\odot} $), suggesting higher inferred densities for low-mass stars under this extinction law.
   - The **RL-EL distribution (blue)** exhibits a broader peak and a heavier tail into higher mass ranges ($>40 \, M_{\odot} $).

2. **Dominance of Low-Mass Stars**:
   - Both curves peak at $<20 \, M_{\odot} $, consistent with the shape of the initial mass function (IMF), where low-mass stars dominate.

3. **Differences in High-Mass Range**:
   - The blue curve (RL-EL) suggests a greater proportion of high-mass stars compared to the orange curve (N-EL), which declines more steeply.

4. **Implications for the IMF**:
   - The choice of extinction law has a **significant impact** on the derived stellar mass distribution.
   - These **variations emphasize** the need for **careful consideration** of extinction corrections when studying the IMF, as **small changes** in extinction assumptions can lead to **notable** differences in the inferred mass function.

### Probability of finding $(\alpha \cdot M_\odot) \pm 0.5\cdot M_\odot$ like stars in subsets of the Arches Cluster (for the RL-EL)
The probability was estimated by bootstrapping a pre-filtered dataset, determining the fraction of stars within a range of $\pm 0.5 \cdot M_\odot$ around $\alpha \cdot M_\odot$ in each sample, and then averaging these probabilities.

```python
# df should be pre filtered
def prob(M_in_solar_masses: float, df, reps: int):
    probabilities = []
    df_values = np.array(df["Mi(RL)"])
    
    for _ in range(reps):
        sample = np.random.choice(df_values, size=len(df_values), replace=True)
        
        probability = np.sum((sample >= M_in_solar_masses - 0.5) & (sample <= M_in_solar_masses + 0.5)) / len(sample)
        probabilities.append(probability)
    
    return probabilities

"The probability of obtaining a star with an inital mass alpha*M_sun in (alpha - 0.5, alpha + 0.5)*M_sun is {:.3f}%".format(np.mean(prob(20.0, filtered_initial_mass_RL_df, 10))*100)
```

### Law of Large Numbers

**Sample Mean**:
    $\displaystyle \overline{X_n} = \frac{1}{n}\sum_{i=1}^n X_i$
**Population Mean**:
    $\displaystyle \mu = E[X]$

The law of large numbers states that 

$\forall \epsilon > 0, \quad \lim_{n \rightarrow \infty} P(|\overline{X_n} - \mu| > \epsilon) = 0$

This suggests that when taking random samples the resulting shape of the distributions should be similar for all sample distributions.

```python
fig, ax = plt.subplots(2,2,figsize=(15,10))

ax[0,0].hist(filtered_initial_mass_RL_df["Mi(RL)"],bins=50, edgecolor='steelblue',color='steelblue',alpha=0.25)

s_1 = filtered_initial_mass_RL_df["Mi(RL)"]

def subsetter(dataframe: pd.DataFrame,n: int):
    random_gen = random.sample([i for i in range(0,len(dataframe))],k=len(dataframe)//n)
    
    s = [dataframe.to_numpy()[index] for index in random_gen]
    
    return s

ax[0,1].hist(subsetter(s_1,2),bins=50, edgecolor='steelblue',color='steelblue',alpha=0.25)

ax[1,0].hist(subsetter(s_1,10),bins=50, edgecolor='steelblue',color='steelblue',alpha=0.25)

ax[1,1].hist(subsetter(s_1,20),bins=50, edgecolor='steelblue',color='steelblue',alpha=0.25)
```

### Initial Mass Function Estimation
Due to the nature of the inital mass distribution we can either chose a power law approach with different slopes for different sections of the IMF or a Log-Normal approach due to the positively skewed distribution of initial masses.

Here we chose Log-Normal apporach.

As a quick reminder:

The **Log-Normal Distribution** is defined as follows:

$$f(x,\sigma) = \frac{1}{x\sqrt{2\pi \sigma^2}}\exp\left(-\frac{(\ln{x} - \mu)^2}{2 \sigma^2}\right)$$


```python
fig, ax = plt.subplots()

ax.hist(filtered_initial_mass_RL_df["Mi(RL)"],bins=100,edgecolor='steelblue',color='steelblue',alpha=0.25)
ax.set_xlim(0,100)
ax.set_ylabel('Frequency')
ax.set_xlabel('M_Sun')


median = filtered_initial_mass_RL_df["Mi(RL)"].median()

mean = (1/len(filtered_initial_mass_RL_df["Mi(RL)"]))*(np.array(filtered_initial_mass_RL_df["Mi(RL)"]).sum())
variance = (1/len(filtered_initial_mass_RL_df["Mi(RL)"]))*((np.array((np.array(filtered_initial_mass_RL_df["Mi(RL)"])-np.array(mean)))**2).sum())

#mean, median
ax.axvline(x=mean,color='red', label=f'mean: {mean:.2f}')
ax.axvline(x=median, color='orange', label=f'median: {median}')

ax.legend()
```

```python
from scipy.stats import lognorm
from scipy.optimize import curve_fit

#filtered_initial_mass_RL_df["Mi(RL)"]
filtered_initial_mass_RL = np.array([e for e in filtered_initial_mass_RL_df["Mi(RL)"]])

shape, loc, scale = lognorm.fit(filtered_initial_mass_RL, floc=0) #

# Generate values for the PDF
x = np.linspace(min(filtered_initial_mass_RL), max(filtered_initial_mass_RL), 1000)
pdf = lognorm.pdf(x, shape, loc, scale)

#Parameters
mu = np.log(scale) #scale = median mu^star = e^mu
sigma = shape

#Confidence interval bounds in lognormal space
z = 1.96
lower = np.exp(mu - z*sigma*(1/np.sqrt(len(filtered_initial_mass_RL))))
upper = np.exp(mu + z*sigma*(1/np.sqrt(len(filtered_initial_mass_RL))))

# Plot histogram and fitted PDF
plt.figure(figsize=(10, 6))
plt.hist(filtered_initial_mass_RL, bins=30, density=True, alpha=0.6, color='gray', edgecolor='black', label='Mi(RL) frequency per M_solar')
plt.plot(x, pdf, 'r-', label=f'Lognormal fit\nShape={shape:.2f}, Scale={scale:.2f}')
plt.axvline(lower, color='blue', linestyle='--', label='lower bound')
plt.axvline(upper, color='green', linestyle='--', label='upper bound')
plt.axvline(scale, color='orange',linestyle='--', label='estimated median')
plt.xlabel('Stellar Mass [M_solar]')
plt.ylabel('Density')
plt.title('Lognormal Fit to inital mass (RL - Law) for the Arches Cluster')
plt.legend()
plt.show()
```

Taking the median $\mu_*=e^\mu$ as the estimator in the log normal space of our lognormal fit. We can get the confidence Interval (95%) for the median through
$$(e^{\mu - z\frac{\sigma^2}{\sqrt{n}}},e^{\mu + z\frac{\sigma^2}{\sqrt{n}}})\;\;\; z=1.96$$
When doing so we estimate $\sigma$ through the shape parameter, given by the lognormal fit from scipy.

There is a slight discrepancy in the **sample median** in comparison to the median obtained by the lognormal fit. This is probably due to the fact that the data might indeed by shaped according to a power law, as the theory for the IMF suggests.

This is noticeable when looking at the 95% confidence interval.

```python
conv_array_ONC_mass = np.array([e for e in df_ONC[df_ONC["MH97b"] != 0.0]["MH97b"]])

shape, loc, scale = lognorm.fit(conv_array_ONC_mass, floc=0) #

# Generate values for the PDF
x = np.linspace(min(conv_array_ONC_mass), max(conv_array_ONC_mass), 1000)
pdf = lognorm.pdf(x, shape, loc, scale)

#Parameters
mu = np.log(scale) #scale = median mu^star = e^mu
sigma = shape

#Confidence interval bounds in lognormal space
z = 1.96
lower = np.exp(mu - z*sigma*(1/np.sqrt(len(conv_array_ONC_mass))))
upper = np.exp(mu + z*sigma*(1/np.sqrt(len(conv_array_ONC_mass))))

# Plot histogram and fitted PDF
plt.figure(figsize=(10, 6))

#density normalizes the bar lengths as a pdf; length is defined as counts per bin
plt.hist(conv_array_ONC_mass, bins=1000, density=True, alpha=0.6, color='gray', edgecolor='black', label='Mi(N) frequency per M_solar')
plt.plot(x, pdf, 'r-', label=f'Lognormal fit\nShape={shape:.2f}, Scale={scale:.2f}')
plt.xlim(0,1)
plt.axvline(lower, color='blue', linestyle='--')
plt.axvline(upper, color='green', linestyle='--')
plt.axvline(scale, color='orange',linestyle='--')
plt.xlabel('Stellar Mass [M_solar]')
plt.ylabel('Density')
plt.title('Lognormal Fit to inital mass (N - Law) for the Arches Cluster')
plt.legend()
plt.show()
```

As can be observed the Log-Normal fit is not ideal. But the yielded estimator for the median (Scale) is close enough to the sample median.

### Mass-Cluster Relationship using K-Means

#### Hypothesis Testing the Mass-Cluster Relationship based on true fields F and predicted Clusters through K-Means

$H_0: Cluster assignments are independent of true source fields F$
$H_a: Cluster assignments are dependent on true source fields$

**Rejection Rule**:
If Purity-Test shows that $\text{Purity} < 1$, we have to assume that K-Means perhaps isn't the best option to cluster our data.

#### Preprocessing the Data
Standardize our data using the scikit implementation of the standardization $$z_i = \frac{x_i - \mu}{\sigma},$$ as is recommended for K-Means.

#### K-Means

Given $(X_1,X_2,...,X_n)$, where $X_i \in \mathbb{R}^d$ the K-Means algorithm selects $k$ initial **centroids**, which are points in the feature space.\
Usually the mean $\mu_i$ of each such subset $S_i \subset (X_1,X_2,...,X_n)$ is called centroid, where 
$$\mu_i = \frac{1}{|S_i|}\sum_{x \in S_i} X$$
with $|S_i|$ being the size of $S_i$.

The goal of K-Means is than to minimize the within-cluster sum of squares (WCSS).
$$\argmin_S \sum_{i=1}^k \sum_{X \in S_i} ||X - \mu_i||^2 = \argmin_S \sum_{i=1}^k |S_i| \text{Var}\;S_i$$

A bit less formaly we can say that the algorithm assigns a meausure to a subset of the dataset, with which we are able to compute the distance of each point to its centroid and than assigns it to the nearest cluster. After doing so we take try to find the cluster with the smallest distance. Repeat this step and one can find the suposed clsuters. 

**References**:
1. **K-Means Clustering**. *Wikipedia*. Available at: [https://en.wikipedia.org/wiki/K-means_clustering](https://en.wikipedia.org/wiki/K-means_clustering) (Accessed on: 4 February 2025).


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaled_df = df[df["Mi(RL)"] != -999]
scaled_df["Radial_Distance"] = np.sqrt(np.array(scaled_df['oDE'])**2 + np.array(scaled_df['oRA'])**2)

# Standardize the features for better clustering performance; leads to negative initial masses
scaler = StandardScaler()
scaled_df[['Mi(RL)', 'Radial_Distance']] = scaler.fit_transform(scaled_df[['Mi(RL)', 'Radial_Distance']])

# Apply K-Means clustering with optimal k (choosing k=3 based on expected segregation)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
scaled_df["KMeans_Cluster"] = kmeans.fit_predict(scaled_df[['Mi(RL)', 'Radial_Distance']])

# Scatter plot of actual data points colored by clusters
plt.scatter(scaled_df["Radial_Distance"], scaled_df["Mi(RL)"], c=scaled_df["KMeans_Cluster"], cmap="viridis", s=10, edgecolors="k")

plt.xlabel("Radial Distance (arcsec)")
plt.ylabel("Present Stellar Mass (M_sun)")
plt.title("K-Means Cluster Segregation in Mass Segregation Analysis")
plt.colorbar(label="Cluster ID")
plt.grid(True)
plt.show()

```

This plot is not on a logarithmic scale, the negative masses and Radial Distances arise from the standardization of the data for the K-Means algorithm. It shows stars with similar present day mass are usually grouped together, which seems to be right according to star formation theory. We can now compare these results derived from the K-Means algorithm to the source Cluster Fields for each star in our dataset to test the viability of K-Means for this sort of study. In literature algorithms such as DBSCAN and HDBSCAN are known to be used to compute such charactaristics.

```python
scaled_df[["KMeans_Cluster","F"]]
```

One can observe that certain data points seem to be mislabeled. We can use some evaluation methods from [Cluster Analysis](https://en.wikipedia.org/wiki/Cluster_analysis) such as the Purity defined as $$\text{Purity} = \frac{1}{N}\sum_k \max_j |C_k \cap F_j|$$, where $N$ is the total number of data points, $C_k$ is the set of data points in a K-Means cluster $k$ and $F_j$ is the set of data points within the label group $j$ (for our dataset).

```python
from sklearn.metrics import confusion_matrix

# confusion matrix
conf_matrix = confusion_matrix(scaled_df["F"], scaled_df["KMeans_Cluster"])

purity = np.sum(np.max(conf_matrix, axis=0)) / len(scaled_df)

print(f"Purity: {purity:.3f}")
```

Based on [Python Clustering 'purity' metric](https://stackoverflow.com/questions/34047540/python-clustering-purity-metric), but using a confusion matrix instead of a contingency matrix ([Difference between a Confusion Matrix and Contigency Table](https://stackoverflow.com/questions/7612982/what-is-the-difference-between-a-confusion-matrix-and-contingency-table)). A Purity of 0.652 suggests not optimal clustering.

This plot shows the application of linear regression to model the relationship between the initial mass of stars (x-axis in $M_\odot $) and their Ks-band magnitude (y-axis, measured in magnitudes). There is a clear trend on the observed data points, which the linear regression aims to quantify. 

The relationship is modeled using the formula:

$$ \mathbf{y = mx + \beta} $$

where:

- \( x \): Initial mass ($M_\odot $), independent variable),
- \( m \): Slope of the regression line,

The intercept ($\beta$ = 15.54) shows the magnitude when the initial mass is zero. However this does not have a meaningful physical interpretation beacause a mass of zero $M_\odot $ is not realistic for stars.

The correlation coefficent ($|r_{XY}|$ = 0.868) confirms a strong linear relationship between variables. 

Physical interpretation: This liear regression shows the relationship of inital mass and Ks-band magnitude, aligning with the expectation that more massive stars are genereally brighter. The modell works good within the range of the observed data, but care must taken not to overinterpret values close to zero or negative masses.

Conclusion: The intercept, while mathematically necessary to define the linear model, does not hold physical significance in this context. This highlights an important limitation of linear regression models: while they effectively describe trends in data, some parameters may not always be physically meaningful. Despite this, the slope and overall trend are consistent with physical theory and provide valuable insights into stellar properties.


```python
def cov(X,Y):
    n = len(X)
    alpha = (1/(n-1))
    res = 0
    if len(X) == len(Y):
        for i in range(0,n):
            res += (X[i]-X.mean())*(Y[i]-Y.mean())
    return alpha*res

def S_X_2(X,Y):
    n = len(X)
    alpha = (1/(n-1))
    res = 0
    if len(X) == len(Y):
        for i in range(0,n):
            res += (X[i]-X.mean())**2
    return alpha*res

def S_Y_2(X,Y):
    n = len(Y)
    alpha = (1/(n-1))
    res = 0
    if len(X) == len(Y):
        for i in range(0,n):
            res += (Y[i]-Y.mean())**2
    return alpha*res

r_XY = lambda X,Y: cov(X,Y)/(S_X_2(X,Y)*S_Y_2(X,Y))
```

### Covariance and Correlation Coefficient Calculation

This section implements functions to calculate the **covariance** and **correlation coefficient** between two variables \( X \) and \( Y \).

#### Functions:


1. **cov(X, Y)**:
   - Calculates the covariance between two datasets \( X \) and \( Y \).
     $$
     \text{Cov}(X, Y) = \frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})
     $$
   - The function iterates over the elements of \( X \) and \( Y \), computing the product of deviations from their respective means.
   - A scaling factor $\frac{1}{n-1}$ is applied to return the unbiased estimate of covariance.


2. **S_X_2(X)** and **S_Y_2(Y)**:
   - Calculate the variance for \( X \) and \( Y \), respectively.
   
     $$
     \text{Var}(X) = \frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2
     $$
   - These functions compute the squared deviations of each value from the mean, then scale by $\frac{1}{n-1}$.


3. **$r_{XY}$**:
   - Computes the correlation coefficient $r_{XY}$, which measures the linear relationship between \( X \) and \( Y \).
     $$
     r_{XY} = \frac{\text{Cov}(X, Y)}{\sqrt{\text{Var}(X) \cdot \text{Var}(Y)}}
     $$
   - This function divides the covariance by the product of the standard deviations of \( X \) and \( Y \) (derived from variances).

#### Results:
- **Covariance**:
  $
  \text{Cov}(X, Y) = -1.372
  $
- **Correlation Coefficient**:
  $
  r_{XY} = -0.632
  $

     



```python
# Bootstrap
def Bootstrap(X,Y,Nboot):
    N = len(Y)
    
    resampled_reg = []
    bootstrap_coeffs = []
    for k in range(Nboot):
        random_ind = np.random.randint(0,len(Y),len(Y))
        X_sample = X[random_ind].reshape((-1,1))
        Y_sample = Y[random_ind]
        model = LinearRegression().fit(X_sample,Y_sample)
        y_pred = model.predict(X_sample)
        resampled_reg.append([X_sample,y_pred])
        bootstrap_coeffs.append([model.intercept_,model.coef_[0]])

    return resampled_reg, bootstrap_coeffs

bootstraped_data = Bootstrap(x,y,1000)[0]

fig, ax1 = plt.subplots()

ax1.scatter(x,y,s=0.3,color='black')
ax1.set_xlabel("Initial Mass in M_sun")
ax1.set_ylabel("Ks-Band magnitude in mag")
ax1.invert_yaxis()
for i in range(0,len(bootstraped_data)):
    ax1.plot(bootstraped_data[i][0],bootstraped_data[i][1],c='grey',lw=0.1)
ax1.plot(x,y_pred,color="blue",lw=0.5)
```

The bootstrap quantifies uncertainty in the regression model by analyzing how the slope and intercept change with different data samples. This is achieved by repeatedly resampling the original dataset with replacement, allowing some points to be repeated while others are excluded. A regression line is fitted to each resampled dataset, and the slope and intercept are recorded. The spread of the bootstrap regression lines visually represents this variability, while confidence intervals for the slope and intercept are derived from the distribution of the resampled coefficients. This method offers a simple and flexible approach to assess the modelâ€™s stability without relying on strict mathematical assumptions.


### Results and Conclusion
The bootstrap results confirm a negative relationship between initial mass and Ks-band magnitude, showing that more massive stars are generally brighter. The spread of the bootstrap regression lines reflects the model's uncertainty, with a narrower spread indicating a more stable regression and a wider spread suggesting greater variability. Confidence intervals for the slope and intercept provide a range where the true values are likely to fall, helping to quantify uncertainty. Overall, the analysis supports the observed trend while accounting for natural variability in the data.



```python
bootstrap_coeffs = np.array(Bootstrap(x,y,200)[1])

conf_int = np.percentile(bootstrap_coeffs,[2.5,97.5], axis=0)

plt.hist(bootstrap_coeffs[:, 1], bins=30, alpha=0.7, color='skyblue',edgecolor='black')
plt.axvline(conf_int[0, 1], color='red', linestyle='--',label='2.5% CI')
plt.axvline(conf_int[1, 1], color='blue', linestyle='--',label='97.5% CI')
plt.legend()
```

### Bootstrap Confidence Intervals for Regression Coefficients

This section demonstrates how bootstrap resampling is used to estimate the confidence intervals for the slope of the regression line.

### Histogram

The histogram shows the distribution of boostrap estimates for the slope of the regression line. The peak marks the point with the most frequently observed slope value during the resampling. The red (lower bound) and blue (upper bound) dotted line helps to visualize the 95% confidence intervall, where the true slope is likely to fall within this range 95% of the time. The spread of the histogram and the size of the confidence intervall reflect the variability in the slope due to sampling unvertainy. 




### Purpose and Insights

The purpose of this analysis is to quantify the uncertainty in the slope of the regression line by using bootstrap resampling to calculate confidence intervals. By repeatedly resampling the data and estimating the regression slope, this method provides a robust, non-parametric way to evaluate the variability in the slope. The 95% confidence interval (CI), derived from the 2.5th and 97.5th percentiles of the bootstrap estimates, represents the range of plausible values for the slope, indicating where the true slope is likely to lie with 95% confidence. The histogram highlights the most frequent values of the slope, while the spread of the distribution reflects the variability due to sampling. A narrower confidence interval suggests greater precision in the slope estimate, while a wider interval indicates more uncertainty. This approach is advantageous because it does not rely on parametric assumptions, making it applicable to a wide variety of datasets.
